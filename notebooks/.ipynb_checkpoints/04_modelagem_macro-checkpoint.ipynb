# Modelagem Preditiva da Inadimpl√™ncia

## Objetivo

Desenvolver um modelo econom√©trico baseado em regress√£o linear para estimar a taxa agregada de inadimpl√™ncia no Brasil a partir de vari√°veis macroecon√¥micas relevantes, especificamente taxa de juros e taxa de desemprego.

O objetivo n√£o √© apenas prever valores futuros, mas tamb√©m compreender e interpretar economicamente o impacto dessas vari√°veis sobre o comportamento da inadimpl√™ncia ao longo do tempo.

üß† Estrat√©gia de Valida√ß√£o
## Estrat√©gia de Valida√ß√£o

Considerando que os dados possuem natureza temporal, n√£o foi utilizada divis√£o aleat√≥ria (train_test_split), a fim de evitar vazamento de informa√ß√£o e preservar a ordem cronol√≥gica.

Foi adotada valida√ß√£o temporal estruturada da seguinte forma:

80% dos per√≠odos mais antigos ‚Üí Treinamento

20% mais recentes ‚Üí Teste

Essa abordagem simula um cen√°rio real de previs√£o futura, garantindo maior credibilidade √† avalia√ß√£o do modelo.

Al√©m do modelo base, foram incorporadas an√°lises complementares para refor√ßar a robustez da modelagem:

Avalia√ß√£o de desempenho com m√©tricas estat√≠sticas (MAE, RMSE e R¬≤)

Inclus√£o de vari√°veis defasadas (lag) para capturar efeitos econ√¥micos retardados

An√°lise de multicolinearidade por meio do VIF

Compara√ß√£o visual entre valores reais e previstos


```python
# ==============================
# IMPORTA√á√ïES
# ==============================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
```

```python
# ==============================
# CARREGAMENTO DA BASE TRATADA
# ==============================

df = pd.read_parquet('../data/processed/dataset_tratado.parquet')

df.head()

```

```python
# ==============================
# ORDENA√á√ÉO CRONOL√ìGICA
# ==============================

df = df.sort_values('data')
df.reset_index(drop=True, inplace=True)

print("Per√≠odo dispon√≠vel na base:")
print(df['data'].min(), "at√©", df['data'].max())
```

```python
# ==============================
# DIVIS√ÉO TEMPORAL
# ==============================

tamanho_treino = int(len(df) * 0.8)

df_train = df.iloc[:tamanho_treino]
df_test = df.iloc[tamanho_treino:]

print("Per√≠odo de Treino:")
print(df_train['data'].min(), "at√©", df_train['data'].max())

print("\nPer√≠odo de Teste:")
print(df_test['data'].min(), "at√©", df_test['data'].max())
```

```python
# ==============================
# DEFINI√á√ÉO DAS VARI√ÅVEIS
# ==============================

X_train = df_train[['juros', 'desemprego']]
y_train = df_train['inadimplencia']

X_test = df_test[['juros', 'desemprego']]
y_test = df_test['inadimplencia']
```

```python
# ==============================
# TREINAMENTO DO MODELO
# ==============================

modelo = LinearRegression()
modelo.fit(X_train, y_train)
```

```python
# ==============================
# AVALIA√á√ÉO DO MODELO
# ==============================

y_pred = modelo.predict(X_test)

r2 = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test, y_pred)

print(f"R¬≤: {r2:.4f}")
print(f"MSE: {mse:.4f}")
print(f"RMSE: {rmse:.4f}")
print(f"MAE: {mae:.4f}")
```

```python
# ==============================
# REAL VS PREVISTO
# ==============================

plt.figure(figsize=(10,5))

plt.plot(df_test['data'], y_test.values, label='Real')
plt.plot(df_test['data'], y_pred, label='Previsto')

plt.title('Inadimpl√™ncia - Real vs Previsto')
plt.xlabel('Data')
plt.ylabel('Inadimpl√™ncia')
plt.legend()

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# ==============================
# COEFICIENTES DO MODELO
# ==============================

coeficientes = pd.DataFrame({
    'Vari√°vel': X_train.columns,
    'Coeficiente': modelo.coef_
})

coeficientes
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

y_pred = modelo.predict(X_test)

mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print("MAE:", mae)
print("RMSE:", rmse)
print("R¬≤:", r2)
```
