# Análise Exploratória de Dados (EDA)

## Objetivo

Explorar as séries temporais de inadimplência, juros e desemprego
para identificar padrões, tendências e correlações.

Etapas:

- Visualização temporal das séries
- Cálculo de correlação
- Heatmap para análise de relação entre variáveis


```python
# ==============================
# IMPORTAÇÃO DE BIBLIOTECAS
# ==============================

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ==============================
# CARREGAMENTO DA BASE TRATADA
# ==============================

df = pd.read_parquet('../data/processed/dataset_tratado.parquet')
df.head()

```

```python
# ==============================
# VISUALIZAÇÃO TEMPORAL
# ==============================

plt.figure(figsize=(12,6))
plt.plot(df['data'], df['inadimplencia'], label='Inadimplência')
plt.plot(df['data'], df['juros'], label='Juros')
plt.plot(df['data'], df['desemprego'], label='Desemprego')

plt.xlabel('Data')
plt.ylabel('Valor')
plt.title('Evolução das Variáveis Macroeconômicas')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# ==============================
# MATRIZ DE CORRELAÇÃO
# ==============================

correlacao = df[['inadimplencia', 'juros', 'desemprego']].corr()
correlacao
```

```python
# ==============================
# HEATMAP DE CORRELAÇÃO
# ==============================

plt.figure(figsize=(6,4))
sns.heatmap(correlacao, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlação entre Variáveis Macroeconômicas')
plt.show()
```

# Conclusão

- As séries mostram tendências distintas ao longo do tempo.
- Correlação positiva entre algumas variáveis sugere influência conjunta.
- Informações extraídas aqui servirão de base para modelagem preditiva e geração de insights.
